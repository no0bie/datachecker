```python
import os
import re

class Beers:
    """
    Beers object:
    - passed (int): Number of successful or warned tests
    - failed (int): Number of failed tests
    - total (int): Total number of checks performed
    - log (str): Contains the log of all the tests, can also be displayed if `True` is passed to logging in functions
    - result (bool): `True` if it passed >= 50% of tests, otherwise `False`
    """

    def __init__(self, passed: int, failed: int, total: int, log: str, result: bool):
        self.passed = passed
        self.failed = failed
        self.total = total
        self.log = log
        self.result = result

dir_path = r"\\10.66.10.6\DailyDataRaw\CryptoHist\airflow\cleaned\intrade\BINANCE"
pattern = r"2023-01-30_00.00.00__2023-01-3.*"

a = {}

for root, dirs, files in os.walk(dir_path):
    for file in files:
        if re.match(pattern, file):
            file_path = os.path.join(root, file)
            
            df = pd.read_csv(file_path)
            missing_percent = df['trade_open'].isna().mean() * 100
            failed = 0
            if missing_percent < 5:
                failed += 1
            if len(df) < 5:
                failed += 1

            a[file_path] = Beers(passed=(2 - failed), total=2, failed=failed, result=((2-failed) >= 1), log="Cool")
        
```
